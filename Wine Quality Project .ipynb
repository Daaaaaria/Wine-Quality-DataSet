## Data Dictionary

Der Datensatz enthält mehrere Variablen (oder Spalten), die verschiedene Attribute des Weins beschreiben. Erklärung der zwölf Variablen, die im Datensatz für die Weinqualität enthalten sind:

- **Feste Säure**: Dies bezieht sich auf die Menge der im Wein enthaltenen nichtflüchtigen Säuren, die üblicherweise in Gramm je Liter (g/l) ausgedrückt werden. Feste Säuren verdampfen nicht leicht und tragen somit zum Gesamtsäuregehalt des Weins bei.
- **Flüchtige Säure**: Damit wird die Menge flüchtiger Säuren im Wein gemessen, die bei hohen Konzentrationen zu unangenehmen Aromen beitragen können. Sie wird in der Regel auch in Gramm pro Liter (g/l) ausgedrückt.
- **Citronensäure**: Citronensäure ist eine schwache organische Säure, die natürlich in Zitrusfrüchten vorkommt. Sie kann zur Säure des Weins beitragen und außerdem die Frische und Komplexität des Geschmacks erhöhen.
- **Restzucker**: Dieser Wert gibt die Zuckermenge an, die nach der Gärung im Wein verbleibt, in der Regel in Gramm pro Liter (g/l). Weine mit einem höheren Restzuckergehalt schmecken tendenziell süßer.
- **Chloride**: Chloride beziehen sich auf die Konzentration von Chloridionen im Wein, die normalerweise in Gramm pro Liter (g/l) gemessen werden. Hohe Chloridgehalte können manchmal auf Kontamination oder schlechte Weinherstellungsverfahren hinweisen.
- **Freies Schwefeldioxid**: Schwefeldioxid (SO2) wird häufig als Konservierungsmittel in der Weinbereitung verwendet, um Oxidation und mikrobiellen Verderb zu verhindern. Der Gehalt an freiem Schwefeldioxid stellt die Menge an SO2 dar, die nicht an andere Verbindungen gebunden ist und somit als Konservierungsmittel zur Verfügung steht.
- **Gesamtschwefeldioxid**: Diese Variable stellt die Gesamtmenge an Schwefeldioxid im Wein dar, einschließlich freier und gebundener Formen. Sie wird in der Regel in ppm (Parts per Million) oder Milligramm pro Liter (mg/L) gemessen.
- **Dichte**: Die Dichte ist ein Maß für die Masse pro Volumeneinheit des Weins und wird typischerweise in Gramm pro Milliliter (g/ml) ausgedrückt. Sie kann Informationen über die Gesamtkonzentration von Feststoffen und gelösten Stoffen im Wein liefern.
- **pH**: Der pH-Wert ist ein Maß für den Säuregehalt oder die Alkalinität des Weins auf einer Skala von 0 bis 14, wobei niedrigere Werte eine höhere Säurekonzentration und höhere Werte eine höhere Alkalinität anzeigen. Der pH-Wert kann den Geschmack, die Stabilität und die mikrobielle Sicherheit des Weins beeinflussen.
- **Sulfate**: Sulfate sind chemische Verbindungen, die zum Geschmack und Aroma von Wein beitragen können. Sie können auch antimikrobielle Eigenschaften haben, die den Wein konservieren.
- **Alkohol**: Diese Variable stellt den Alkoholgehalt des Weins dar, der in der Regel als Volumenanteil ausgedrückt wird. Der Alkoholgehalt kann den Körper, den Geschmack und das Mundgefühl des Weines beeinflussen.
- **Qualität**: Dies ist die Zielvariable im Datensatz, die die allgemeine Qualitätseinstufung des Weins darstellt. Die Qualität wird häufig von erfahrenen Weinverkostern bewertet und in der Regel auf einer Skala von 3 bis 9 bewertet, wobei höhere Werte eine bessere Qualität bedeuten.

4. ***Welcher Zusammenhang besteht zwischen einer bestimmten Weineigenschaft und der Qualitätsbewertung eines Weins?***

Um den Zusammenhang zwischen den Merkmalen eines Weins und seiner Qualitätsbewertung zu verstehen, haben wir Boxplots für jedes Merkmal in Bezug auf die Qualität erstellt.

```python
# Create a grid of 4x3 subplots

fig, axs = plt.subplots(4,3, figsize=(18, 20))

# List of variables

variables= ['fixed acidity',
            'volatile acidity',
            'citric acid',
            'residual sugar',
            'chlorides',
            'free sulfur dioxide',
            'total sulfur dioxide',
            'density',
            'pH',
            'sulphates',
            'alcohol',
            'quality'
           ]

# Iterate over the variables and create box plots

for i,variable in enumerate(variables):
    
    row = i // 3
    col = i % 3
    
    sns.boxplot(x='quality', y=variable, data=wine_quality_df, ax=axs[row, col],palette='coolwarm')
    axs[row, col].set_title(f'Quality vs {variable}')
    axs[row, col].set_xlabel('Quality')
    axs[row, col].set_ylabel(variable)

# Adjust layout to prevent overlapping   
    
plt.tight_layout()

# Display the figure

plt.show()
```

Mithilfe der Grafiken können wir den allgemeinen Trend des Einflusses der einen oder anderen Variablen auf die Qualität erkennen, aber aufgrund der großen Variabilität der Qualitätsbewertungen ist es schwierig, ihn nachzuvollziehen.

Um die Klarheit der Grafiken zu erhöhen und die weitere Analyse zu vereinfachen, werden wir die Bewertung der Weinqualität in Untergruppen unterteilen. Dies ist sinnvoll, da es sich bei der Qualitätsbewertung um eine kategoriale Variable handelt. Erstellen wir mithilfe der Funktion *„Where“* eine neue Spalte mit Unterkategorien der Weinqualität.  

Teilen wir den Wein in 3 Unterkategorien ein: Qualität 7,8 – 'good', 5-6 – 'medium', 3,4 – 'bad'

```python
#Creating a column 'quality rating' dividing wine quality into subcategories

wine_quality_df['quality rating']=np.where(
wine_quality_df['quality'] >6,'good',
np.where(wine_quality_df['quality'] <=4,'bad','medium')   
)
```

```python
# Create a grid of 4x3 subplots

fig, axs = plt.subplots(4,3, figsize=(15, 20))

# List of variables

variables= ['fixed acidity',
            'volatile acidity',
            'citric acid',
            'residual sugar',
            'chlorides',
            'free sulfur dioxide',
            'total sulfur dioxide',
            'density',
            'pH',
            'sulphates',
            'alcohol',
            'quality'
           ]

# Iterate over the variables and create box plots

for i,variable in enumerate(variables):
    
    row = i // 3
    col = i % 3
    
    sns.boxplot(x='quality rating', y=variable, data=wine_quality_df, ax=axs[row, col],palette='coolwarm',order=['bad','medium','good'])
    axs[row, col].set_title(f'Quality vs {variable}')
    axs[row, col].set_xlabel('Quality')
    axs[row, col].set_ylabel(variable)

# Adjust layout to prevent overlapping   
    
plt.tight_layout()

# Display the figure

plt.show()
```

5. ***Wie hoch sind die Durchschnittswerte der Merkmale in den einzelnen Qualitätsuntergruppen?***

Berechnen wir die Durchschnittswerte der Variablen in jeder ihrer Untergruppen mithilfe der 'mean-Funktion':

```python
#Arrange the subcategories in a convenient order

quality_order = ['bad', 'medium', 'good']
wine_quality_df['quality rating'] = pd.Categorical(wine_quality_df['quality rating'], categories=quality_order, ordered=True)

#Let's calculate the average for each group

quality_groups = wine_quality_df.groupby('quality rating')
group_mean = quality_groups.mean().round(2).sort_values(by='quality rating')
group_mean
```

```python
#Visualize the results with barplot

group_mean.plot(kind='bar', figsize=(16, 8))

plt.title('Average value by quality groups',size=16)
plt.xlabel('Quality',size=14)
plt.ylabel('Average',size=14)
plt.xticks(rotation=45)

# Display the figure

plt.show()
```

6. ***In welcher Beziehungen stehen Alkohol, Sulfate und Zitronensäure zueinander?***

Alkohol, Sulfate und Zitronensäure zeigten die höchsten positiven Korrelationswerte mit der Weinqualität.  Wir werden versuchen herauszufinden, welche Beziehungen diese Variablen untereinander und auch zur Qualität haben. Wir werden *Scatterplots* erstellen, um Beziehungen zu identifizieren. Die lineare Regressionslinien wird bei einem 95%-Konfidenzintervall dargestellt.

```python
#Set the size of the figure

plt.figure(figsize=(10, 6))

#Visualize the relationships between variables using scatterplot 

sns.scatterplot(
    x='sulphates', y='citric acid', 
    hue='quality rating', palette='viridis',s=80,
    data=wine_quality_df,alpha=0.5, edgecolor=None
)

# Add a regression line

sns.regplot(
    x='sulphates', y='citric acid', 
    data=wine_quality_df, scatter=False, color='green',ci=95
)

# Customize the plot

plt.title('Comparing citric.acid vs sulphates by quality', fontsize=16)
plt.xlabel('sulphates', fontsize=14)
plt.ylabel('citric acid', fontsize=14)
plt.legend(bbox_to_anchor=(1, 1),loc=2)
plt.tight_layout()

# Show the plot

plt.show()
```

<span style="color:green"> Wir können daraus schließen, dass Zitronensäure eine starke lineare positive Beziehung zum Sulfat-Attribut hat.

```python
#Set the size of the figure

plt.figure(figsize=(10, 6))

#Visualize the relationships between variables using scatterplot 

sns.scatterplot(
    x='citric acid', y='alcohol', 
    hue='quality rating', palette='viridis',s=80,
    data=wine_quality_df,alpha=0.5, edgecolor=None
)

# Add a regression line

sns.regplot(
    x='citric acid', y='alcohol', 
    data=wine_quality_df, scatter=False, color='green',ci=95
)

# Customize the plot

plt.title('Comparing alcohol vs citric.acid by quality', fontsize=16)
plt.xlabel('citric acid', fontsize=14)
plt.ylabel('alcohol', fontsize=14)
plt.legend(bbox_to_anchor=(1,1), loc=2)
plt.tight_layout()

# Show the plot

plt.show()
```

<span style="color:green"> Wir können daraus schließen, dass Alkohol einen linearen positiven Zusammenhang mit dem Attribut Zitronensäure hat, wir beobachten jedoch keinen starken Zusammenhang.

<span style="color:green"> Wir können daraus schließen, dass Sulphates einen linearen positiven Zusammenhang mit dem Alcohol hat, wir beobachten jedoch keinen starken Zusammenhang.

Wir haben die Beziehungen zwischen den drei untersuchten Variablen untersucht und können daraus schließen, dass nur Sulfate und Zitronensäure eine starke positive Korrelation aufweisen. Wir wissen bereits, dass Alkohol auch den stärksten positiven Zusammenhang mit der Qualität aufweist. Schauen wir uns also vier Variablen an: Sulfate, Zitronensäure VS Alkohol, Qualität.

```python
#Set the size of the figure

plt.figure(figsize=(10, 6))

#Visualize the relationships between variables using scatterplot 

sns.scatterplot(
    x='sulphates', y='citric acid', 
    size='alcohol', sizes=(20, 200), 
    hue='quality rating', palette='viridis', 
    data=wine_quality_df,alpha=0.5, edgecolor=None
)

# Add a regression line
sns.regplot(
    x='sulphates', y='citric acid', 
    data=wine_quality_df, scatter=False, color='green'
)

# Customize the plot
plt.title('Comparing citric.acid vs sulphates by quality & alcohol', fontsize=16)
plt.xlabel('sulphates', fontsize=14)
plt.ylabel('citric acid', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.tight_layout()

# Show the plot
plt.show()
```
